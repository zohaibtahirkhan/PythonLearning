```python
from matplotlib import pyplot as plt
```

```python
x = [20,21,22,23,24,25]
y = [100000, 150000, 250000, 300000, 350000, 4000000]
z = [530000, 200000, 210000, 220000, 230000, 2400000]
```

## Note
#### You can either add color, grid, markers, linestyle etc. yourself or, you can use some styles using function
#### *print(plt.style.available)* : to see availible styles
#### *plt.style.use("style_name")* : to use a style

```python
plt.plot(x, y, color='#232323', marker="o", linestyle="--", label="Python")
plt.plot(x, z, color='#fc23da', marker="D", label="Devs")
plt.xlabel("Age") 
plt.ylabel("Salaries") 
plt.title("Age and Salary Relation")
plt.grid(True)
plt.legend()
```

To save a fig, use *plt.savefig("plot.png")*

```python
import numpy as np
x_np = np.arange(len(x))
y_np = np.arange(len(y))
z_np = np.arange(len(z))
```

```python
width = 0.3
plt.bar(x_np - width, y_np, width=width)
plt.bar(x_np, z_np, width=width)
plt.xticks(ticks= x_np, label= x)
plt.xlabel("Age") 
plt.ylabel("Salaries") 
plt.title("Age and Salary Relation")
#print(plt.style.available)
plt.style.use('_mpl-gallery')
```
